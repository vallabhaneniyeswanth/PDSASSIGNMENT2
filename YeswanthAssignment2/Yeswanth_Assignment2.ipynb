#b) Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”). (4 points)

```python
load_dataset['Mileage'] = load_dataset['Mileage'].str.extract(r'(\d+\.?\d*)')[0].astype(float)
load_dataset['Engine'] = load_dataset['Engine'].str.replace(' CC', '', regex=False).astype(float)
load_dataset['Power'] = load_dataset['Power'].str.replace(' bhp', '', regex=False).astype(float)
```

```python
load_dataset['New_Price'] = pd.to_numeric(load_dataset['New_Price'].str.replace(' lakh', '', regex=False), errors='coerce')
```

#d) Create one more feature and add this column to the dataset (you can use mutate function in R for this). For example, you can calculate the current age of the car by subtracting “Year” value from the current year.   (4 points)

we can categorize cars into mileage groups (e.g., Low, Medium, High) based on their mileage

```python
# Define mileage range
ranges = [0, 15, 20, 25, float('inf')]
labels = ['Low', 'Medium', 'High', 'Very High']

load_dataset['Mileage_Category'] = pd.cut(load_dataset['Mileage'], bins=ranges, labels=labels)
load_dataset[['Mileage','Mileage_Category']].head(10)
```

```python
# Save the updated dataset to a CSV file
load_dataset.to_csv('FeatureAdded.csv', index=False)
```

#e) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset. (4 points)

1.select

```python
cols = load_dataset[['Name', 'Year', 'Mileage', 'Price','Owner_Type','Price','Seats','Power','Mileage_Category']] #select operation
cols
```

2.Filter

```python
rows_filetered = load_dataset[(load_dataset['Owner_Type'] == 'First') & (load_dataset['Price'] > 25)]#Filtering cars that are 'First' owner type and have a price greater than 25
rows_filetered
```

3.Rename columns

```python
cols = load_dataset.rename(columns=lambda x: {'Power': 'Horsepower', 'Seats': 'Num_Seats'}.get(x, x))
cols
```

4.mutate

```python
cols['Horsepower'] = cols['Horsepower'] * 1.34102  # 1 kW = 1.34102 BHP
cols['Horsepower']
```

5.Arrange

```python
cols = cols.sort_values(by=['Mileage', 'Year'], ascending=[False, True]) #Arrange by Mileage (descending) and Year (ascending)
cols[['Mileage','Year']]
```

6.Summarize

```python
# Group by 'Owner_Type' and calculating  mean price and mileage
summarized_data = cols.groupby('Owner_Type').agg(
    Average_Price=('Price', 'mean'),
    Average_Mileage=('Mileage', 'mean')
).reset_index()
summarized_data
```

```python
# Save the updated dataset to a CSV file
cols.to_csv('Operations.csv', index=False)
```
